```python
import numpy as np
import matplotlib.pyplot as plt
import os
from subprocess import run, PIPE, TimeoutExpired
import pickle
```

```python
nbfolder = get_ipython().starting_dir
def path(*xs):
    return os.path.join(nbfolder, '..', *xs)
```

```python
with open(path('saved', executable, 'all.pickle'), 'rb') as f:
    param_res = pickle.load(f)
```

```python
for k, v in enumerate(param_res):
    _dimensions = np.array(v['dimensions'])
    _times = np.array(v['times'])
    _order = np.argsort(_dimensions)
    _dimensions = _dimensions[_order]
    _times = _times[_order]

    plt.plot(
    #     [_dimensions[i] for i in range(len(_times)) if _times[i]>1], 
    #     [_times[i] for i in range(len(_times)) if _times[i]>1], 
        _dimensions,
        _times/1000**2,
        'o'
    )
    plt.xlabel('cantidad de pntos')
    plt.ylabel('Time(ms)')
    plt.title(executable)
    plt.savefig(path('saved', executable+str(k)+'.svg'))
    plt.show()
```
