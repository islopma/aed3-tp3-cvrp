```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10
r = 5

indices = np.array(range(n))
np.random.shuffle(indices)

angles = [2*np.pi*i/n for i in indices]
xs = r*np.sin(angles)
ys = r*np.cos(angles)
```

```python
def plot():
    plt.plot([xs[-1], *xs], [ys[-1], *ys])
```

```python
plot()
```

```python
def swap(i, j):
    assert i < j
    while j > i:
        xs[i], xs[j] = xs[j], xs[i]
        ys[i], ys[j] = ys[j], ys[i]
        i+=1
        j-=1
```

```python
def ds(i, j):
    return abs(xs[i]-xs[j]+1j*(ys[i]-ys[j]))
```

```python
def fix():
    for i in range(n):
        for j in range(i):
            if ds(i, (i+1)%n)+ds(j, (j-1)%n) > ds(j, (i+1)%n)+ds(i, (j-1)%n):
                swap(j, i)
                return
```

```python
fix(); plot()
```
