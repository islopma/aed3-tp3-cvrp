```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
r = 10
angles = np.pi/2 + np.array([-0.3, -0.2, -0.05, 0.05, 0.2, 0.3])
```

```python
xs = np.concatenate((r*np.sin(angles), [0]))
ys = np.concatenate((r*np.cos(angles), [0]))
```

```python
s = np.hypot(xs[0]-xs[1], ys[0]-ys[1] )
b = np.hypot(xs[1]-xs[2], ys[1]-ys[2] )
s, b
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(122)
plt.axis('equal')
for path in [[-1,0,1,2,-1], [-1,3,4,5,-1]]:
    plt.plot(xs[path], ys[path], 'o-')

plt.xlabel('Optima (distancia = {:.2f})'.format(4*r+2*s+2*b))

plt.subplot(121)
plt.axis('equal')
for path in [[-1,0,1,-1], [-1,2,3,-1], [-1,4,5,-1]]:
    plt.plot(xs[path], ys[path], 'o-')

plt.xlabel('Savings (distancia = {:2f})'.format(6*r+3*s))
```

```python
for x, y in zip(xs, ys):
    print(x, y)
```
